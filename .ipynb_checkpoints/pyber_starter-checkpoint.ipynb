```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"
```

```python
# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)
# Combine the data into a single dataset
merge_pyber_df = pd.merge(ride_data, city_data, how='left', on='city')
# Display the data table for preview
merge_pyber_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban_city_type = merge_pyber_df[merge_pyber_df["type"] == "Urban"].groupby([merge_pyber_df["city"]])
rural_city_type = merge_pyber_df[merge_pyber_df["type"] == "Rural"].groupby([merge_pyber_df["city"]])
suburban_city_type = merge_pyber_df[merge_pyber_df["type"] == "Suburban"].groupby([merge_pyber_df["city"]])

# Build the scatter plots for each city types
x_ride_urban = urban_city_type["ride_id"].count()
y_fare_urban = urban_city_type["fare"].mean()
s_driver_count_urban = urban_city_type["driver_count"].mean()

x_ride_rural = rural_city_type["ride_id"].count()
y_fare_rural = rural_city_type["fare"].mean()
s_driver_count_rural = rural_city_type["driver_count"].mean()

x_ride_suburban = suburban_city_type["ride_id"].count()
y_fare_suburban = suburban_city_type["fare"].mean()
s_driver_count_suburban = suburban_city_type["driver_count"].mean()

# Incorporate the other graph properties
plt.scatter(x_ride_urban, y_fare_urban, label = "Urban", s=s_driver_count_urban * 10, color=["coral"], edgecolor="black", alpha = 0.75, marker="o")
plt.scatter(x_ride_rural, y_fare_rural, label = "Rural", s=s_driver_count_rural * 10, color=["gold"], edgecolor="black", alpha = 0.75, marker="o")
plt.scatter(x_ride_suburban, y_fare_suburban, label = "Suburban", s=s_driver_count_suburban * 10, color=["skyblue"], edgecolor="black", alpha = 0.75, marker="o")
plt.grid()

# Creating the legend
legend = plt.legend(fontsize = 10, title= "City Types", loc="best") 

# Adjusting the legend 
legend.legendHandles[0]._sizes = [50]
legend.legendHandles[1]._sizes = [50]
legend.legendHandles[2]._sizes = [50]

# Incorporate a text label regarding circle size
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.title("Pyber Ride Sharing Data (2018)")
plt.text(45,30,"Note: \nCircle size correlates with driver count per city.", fontsize = 10)

# Show Figure
plt.show()
```

## Total Fares by City Type

```python
# Calculating Total Fares by City Type in Percents
type_grouped = merge_pyber_df.groupby(['type'])
fare_sum = type_grouped['fare'].sum()

# Building Pie Chart
labels = ["Rural","Suburban","Urban"]
explode = (0, 0, 0.1)
colors = ["yellow", "lightgreen", "orange"]
plt.pie(fare_sum, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Fares by City Type")

# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculating Total Rides by City Type Percents
type_grouped = merge_pyber_df.groupby(['type'])
Rides_count = type_grouped['ride_id'].count()

# Building Pie Chart
labels = ["Rural","Suburban","Urban"]
explode = (0, 0, 0.1)
colors = ["yellow", "lightgreen", "grey"]
plt.pie(Rides_count, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Rides by City Type")

# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculating The Toatl Drivers by City Type Percents
type_grouped_drivers = city_data.groupby(['type'])
Drivers_count = type_grouped_drivers['driver_count'].sum()

# Building Pie Chart
labels = ["Rural","Suburban","Urban"]
explode = (0, 0, 0.1)
colors = ["yellow", "lightgreen", "lightpink"]
plt.pie(Drivers_count, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=150)
plt.title("% of Total Divers Count by City Type")

# Show Figure
plt.show()
```
